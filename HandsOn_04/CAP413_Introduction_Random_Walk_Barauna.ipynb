# 0 - Introduction
## The Random Walk 

The random walk is also called the drunk's walk. In this algorithm we will have the condition that at a certain point we will take a step forward or backward so that the probability of taking a step forward is the same as taking a step back. This motion is also called Brownian motion and has many applications in physics. Here we will model this problem for in one dimension, two dimensions, three dimensions and four dimensions.

# 1 - First Case: 1 Dimensioanl Walk 

If a given subject is on a street that has no exit, this street is of infinite size and he is in position 0 and can move in a positive direction and we can move to a negative position on this street, this subject then gives a step forward and then decide whether to stay in position or step forward or step back this is the random walk algorithm.

Let's suppose then that this guy is at Dinha's bar on the edge of Salvador in the neighborhood of "Rio Vermelho".

## 1.1 Singel and Sad drunk

```python
size, param = 1000, 0 
time = np.arange(size) #Time Vector
steps_x = []

for i in range(size): 
  steps_x.append(np.random.choice([-1,0,1]) + param)
  param = steps_x[i]
   
figure(figsize=(17, 6), dpi=80)
plt.plot(time,steps_x,label='Drunk One',color='blue')
plt.axhline(y =0, color = 'r', linestyle = '--', label='Bar Location')
plt.xlabel('Times of stepes')
plt.ylabel('Position of the Step')
plt.title(f'The Random Walk for {size} steps')
plt.legend()
plt.show()
```

```python
param_1 = param_2=param_3=param_4=param_5 = 0 
steps_1,steps_2,steps_3,steps_4,steps_5, = [],[],[],[],[]

for i in range(size): 
  steps_1.append(np.random.choice([-1,0,1]) + param_1)
  steps_2.append(np.random.choice([-1,0,1]) + param_2)
  steps_3.append(np.random.choice([-1,0,1]) + param_3)
  steps_4.append(np.random.choice([-1,0,1]) + param_4)
  steps_5.append(np.random.choice([-1,0,1]) + param_5)
  param_1 = steps_1[i]
  param_2 = steps_2[i]
  param_3 = steps_3[i]
  param_4 = steps_4[i]
  param_5 = steps_5[i]

#Chart Configuration   
figure(figsize=(17, 6), dpi=80)
plt.plot(time,steps_1,label='Drunk One',color='blue')
plt.plot(time,steps_2,label='Drunk Two',color='red')
plt.plot(time,steps_3,label='Drunk Tree',color='green')
plt.plot(time,steps_4,label='Drunk Four',color='yellow')
plt.plot(time,steps_5,label='Drunk Five',color='pink')
plt.axhline(y =0, color = 'r', linestyle = '--', label='Bar Location')
plt.xlabel('Times of stepes')
plt.ylabel('Position of the Step')
plt.title(f'The Random Walk for {size} steps')
plt.legend()
plt.show()
```
