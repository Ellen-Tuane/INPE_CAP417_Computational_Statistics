# Linear Regression

Main Objectives: 

*   Establish if there is a relationship between two variables: If one increases, does the other one icreaseas too? In others words, are those two variables directly proporcional? 
Exemples: Income and spending, wage and gender, student height and exam score.
*   Forecast new observations: discover about relationship to forecast unobserved values.

Variables:

*    Dependent Variable Y: Explore value, explain or forecast.
*    Independent Variable X: Explain the dependent variable.

Equation:

*   y = ax + b
*   a = slope, coeficiente, weight
*   b = interception, bias
*   x = feature, atribute


# 1) Perfect Linear Relation

*  a < 0, a = 0, a > 0
*  b < 0, b = 0, b > 0

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from numpy.random import seed
from numpy.random import randint
import random
from numpy.polynomial.polynomial import polyfit
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.datasets import load_boston
from sklearn.datasets import load_iris
from mlxtend.feature_selection import ExhaustiveFeatureSelector as EFS

iris = load_iris()
X = iris.data
y = iris.target

lr = LinearRegression()

efs = EFS(lr, 
          min_features=1,
          max_features=4,
          scoring='neg_mean_squared_error',
          cv=10)

feature_names = ('sepal length', 'sepal width', 'petal length', 'petal width')
efs = efs.fit(X, y, custom_feature_names=feature_names)
```

```python
df = pd.DataFrame.from_dict(efs.get_metric_dict()).T
df.sort_values('avg_score', inplace=True, ascending=False)
df
```

```python
df.loc[:, "std_err"] = df["std_err"].map('{:.3f}'.format)
```

```python
df.loc[:, "avg_score"] = df["avg_score"].map('{:.3f}'.format)
```

```python
df = df.astype(str)
```

```python
plt.plot(df["feature_names"], df["avg_score"], 'o', color='red')
plt.plot(df["feature_names"], df["avg_score"], '-', color='green')
plt.xlabel("Atributes")
plt.ylabel("Score")
plt.title("Average score Analysis")
plt.xticks(rotation=90)

plt.figure(figsize=(30, 100))
plt.show()
```

```python
plt.plot(df["feature_names"], df["std_err"], 'o', color='red')
plt.plot(df["feature_names"], df["std_err"], '-', color='blue')
plt.xlabel("Atributes")
plt.ylabel("Error")
plt.title("Average Error Analysis")
plt.xticks(rotation=90)

plt.figure(figsize=(30, 100))
plt.show()
```

```python
print('Best subset (corresponding names):', efs.best_feature_names_)
print('Best MSE score: %.2f' % efs.best_score_ * (-1))
print('Best subset:', efs.best_idx_)
```

# 3) Noise

Noise: An invalid signal overlapping valid data. Fundamentally means, some of our data objects have invalid values in some atributes. 

y(x) = ax + b + N



```python
n = 20
g = 0.8
a, b = 3, 5

seed = 48

np.random.seed(seed)
x = np.random.rand(n)
y = a*x+b

#noise added
# np.random.rand(n): 20 values between 0 and 1, mean 0.5
y_noise = a*x + b + g*np.random.rand(n)
y_noise2 = a*x + b * (g*np.random.rand(n))
```

```python
plt.scatter(x, y, c=y, cmap='Blues')
plt.scatter(x, y_noise, c=y, cmap='autumn')
plt.scatter(x, y_noise2, c=y, cmap='summer')
#plt.plot(xa, ya, c='r', ls='-')
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Noise Added")
plt.show()
```

```python
vm = 0 
sdev = 500
points = 100
y_list = []
x_noise = np.arange(0.0, 100, 1)

for i in x_noise:
    y = 2*i+5
    y_list.append(y)

noise = np.random.normal(loc=vm, scale=sdev, size=points)
noise_f = noise + y_list
```

```python
np.mean(noise)
```

```python
np.mean(noise_f)
```

```python
plt.figure()
plt.plot(noise_f, label="noise n curve",color='red')
plt.plot(noise, label='noise', color='green')
plt.plot(x_noise,y_list, label= 'curve')
plt.legend()
plt.show
```

# Time Series Data 

Multiplicative Noise

Level: The average value in the series.

Trend: The increasing or decreasing value in the series.

Seasonality: The repeating short-term cycle in the series.

Noise: The random variation in the series.

Additive: 
y(t) = Level + Trend + Seasonality + Noise

Multiplicative: 
y(t) = Level * Trend * Seasonality * Noise


```python
from random import randrange
from matplotlib import pyplot
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
series = pd.read_csv("AirPassengers.csv")
# Parse strings to datetime type
series['Month'] = pd.to_datetime(series['Month'], infer_datetime_format=True)
indexedDataset = series.set_index(['Month'])
```

```python
plt.xlabel("Date")
plt.ylabel("Number of air passengers")
plt.plot(indexedDataset, color='red')
```
